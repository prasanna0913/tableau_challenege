```python
import pandas as pd
from functools import reduce
import datetime as dt
import os
```

```python
resource_path = './Resources'
only_files = [os.path.join(resource_path, f) for f in os.listdir(resource_path) if os.path.isfile(os.path.join(resource_path, f)) 
              and f.find('.csv') != -1]
dfs = []
for f in only_files:
    df = pd.read_csv(f)
    dfs.append(df)
citibike_combo = pd.concat(dfs)
citibike_combo.head()
citibike_combo.to_csv('citibike_combo_data.csv', sep = ',')
citibike_combo
```
